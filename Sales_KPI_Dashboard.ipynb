# üìä Sales KPI Dashboard
Interactive Business Intelligence dashboard for monitoring sales performance, revenue trends, and regional growth.

### üóùÔ∏è Key Performance Indicators:
- **Total Revenue**: Comprehensive sales volume tracking.
- **Growth Rate**: Month-over-month performance comparison.
- **Regional Insights**: performance breakdown by geography.
- **Product Analytics**: Identifying top-selling items.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timedelta

# Simulated Sales Data
np.random.seed(101)
dates = [datetime(2023, 1, 1) + timedelta(days=i) for i in range(365)]
regions = ['West', 'East', 'North', 'South']
categories = ['Electronics', 'Clothing', 'Home', 'Beauty']

data = pd.DataFrame({
    'Date': np.random.choice(dates, 1000),
    'Region': np.random.choice(regions, 1000),
    'Category': np.random.choice(categories, 1000),
    'Sales': np.random.uniform(50, 2000, 1000),
    'Profit': np.random.uniform(5, 500, 1000)
})

data['Month'] = data['Date'].dt.strftime('%Y-%m')
print(f"Total Transactions: {len(data)}")
data.head()
```

### üíπ 1. Revenue Trend Analysis

```python
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()
fig = px.line(monthly_sales, x='Month', y='Sales', title='Monthly Revenue Trend', 
              markers=True, line_shape='spline')
fig.update_layout(template='plotly_dark')
fig.show()
```

### üó∫Ô∏è 2. Sales by Region & Category

```python
fig = px.treemap(data, path=['Region', 'Category'], values='Sales', 
                 title='Sales Composition by Region and Category',
                 color='Sales', color_continuous_scale='RdBu')
fig.update_layout(template='plotly_dark')
fig.show()
```
